```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import pickle
from sklearn.metrics import det_curve, roc_curve, roc_auc_score
```

# Ethnicity

```python
with open('../results/vacc_ethnicity_all_cmu_finetuned.pkl', 'rb') as f:
        loaded_dict_g_cmu = pickle.load(f)
loaded_list_g_cmu = []
for k in  loaded_dict_g_cmu.keys():
    loaded_list_g_cmu.append(loaded_dict_g_cmu[k])
std_gar_cmu = []
far = []
for key in loaded_list_g_cmu[0].keys():
    temp = []
    for i in range(len(loaded_list_g_cmu)):
        temp.append(loaded_list_g_cmu[i][key])
    std_gar_cmu.append(np.std(temp, ddof=1)*100)
    far.append(key*100)
```

```python
with open('../results/vacc_ethnicity_all_synth_finetuned.pkl', 'rb') as f:
        loaded_dict_g_synth = pickle.load(f)
loaded_list_g_synth = []
for k in  loaded_dict_g_synth.keys():
    loaded_list_g_synth.append(loaded_dict_g_synth[k])
std_gar_gan = []
for key in loaded_list_g_synth[0].keys():
    temp = []
    for i in range(len(loaded_list_g_synth)):
        temp.append(loaded_list_g_synth[i][key])
    std_gar_gan.append(np.std(temp, ddof=1)*100)
```

```python
fig = plt.figure(figsize=(10, 10))

plt.xscale("log")
widths = np.diff(far + [far[-1] * 10])
plt.bar(x, std_gar_cmu, widths/10, align='edge', facecolor='dodgerblue', edgecolor='white', lw=2, label = 'CMU Multi-Pie')
for i in range(len(std_gar_cmu[0:4])):
    plt.annotate(str(std_gar_cmu[i]), xy=(far[i]+0.001,std_gar_cmu[i]), ha='left', va='bottom')
plt.bar(x+widths/10, std_gar_gan, widths/5, align='edge', facecolor='green', edgecolor='white', lw=2, label = 'Synthetic Faces')
for i in range(len(std_gar_cmu[0:4])):
    plt.annotate(str(std_gar_gan[i]), xy=(far[i]+widths[i]/8,std_gar_gan[i]), ha='left', va='bottom')
plt.xlabel("FAR(%)")
plt.ylabel("GAR Standard Deviation")
plt.title("FAR Vs Std(GAR) - Ethnicity")
plt.legend()
plt.show()
```
